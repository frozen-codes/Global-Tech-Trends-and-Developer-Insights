### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
import pandas as pd
from collections import Counter
# Drop missing values and split language strings
worked_langs = df['LanguageHaveWorkedWith'].dropna().str.split(';')
wanted_langs = df['LanguageWantToWorkWith'].dropna().str.split(';')

# Flatten the lists
worked_flat = [lang.strip() for sublist in worked_langs for lang in sublist]
wanted_flat = [lang.strip() for sublist in wanted_langs for lang in sublist]

# Count frequency
worked_counter = Counter(worked_flat)
wanted_counter = Counter(wanted_flat)

# Create a DataFrame
language_df = pd.DataFrame({
    'WorkedWith': dict(worked_counter),
    'WantToWorkWith': dict(wanted_counter)
}).fillna(0).astype(int)

language_df = language_df.sort_values(by='WorkedWith', ascending=False)
```

```python
import matplotlib.pyplot as plt

# Plot top 10 languages
top_n = 10
top_languages = language_df.head(top_n)

plt.figure(figsize=(12, 6))
bar_width = 0.35
index = range(len(top_languages))

plt.bar(index, top_languages['WorkedWith'], bar_width, label='Worked With')
plt.bar([i + bar_width for i in index], top_languages['WantToWorkWith'], bar_width, label='Want to Work With')

plt.xlabel('Programming Languages')
plt.ylabel('Frequency')
plt.title('Comparison of Programming Languages: Worked With vs Want to Work With')
plt.xticks([i + bar_width / 2 for i in index], top_languages.index, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


```python
!pip install matplotlib_venn
from matplotlib_venn import venn2
import matplotlib.pyplot as plt

# Unique language sets
set_worked = set(worked_flat)
set_wanted = set(wanted_flat)

plt.figure(figsize=(8, 6))
venn2([set_worked, set_wanted], set_labels=('Worked With', 'Want To Work With'))
plt.title("Overlap of Programming Languages")
plt.show()
```

- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
remote_region_df = pd.crosstab(df['Country'], df['RemoteWork'])
top_countries = df['Country'].value_counts().head(10).index
plot_data = remote_region_df.loc[top_countries]

# Plot
plot_data.plot(kind='bar', figsize=(12, 6))
plt.title('Remote Work Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.legend(title='Remote Work Status')
plt.tight_layout()
plt.show()
```

```python
# Normalize values by row for better heatmap visibility (optional)
plot_data_percent = plot_data.div(plot_data.sum(axis=1), axis=0)

# Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(plot_data_percent, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Proportion of Remote Work Status by Country')
plt.xlabel('Remote Work')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 51
})
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop missing values in either column
df_corr = df[['JobSat', 'YearsCodePro']].dropna()
```

```python
satisfaction_map = {
    'Very dissatisfied': 1,
    'Dissatisfied': 2,
    'Neither satisfied nor dissatisfied': 3,
    'Satisfied': 4,
    'Very satisfied': 5
}
df_corr['JobSat_Score'] = df_corr['JobSat'].map(satisfaction_map)
```

```python
pearson_corr = df_corr['JobSat_Score'].corr(df_corr['YearsCodePro'], method='pearson')
print("Pearson Correlation:", pearson_corr)
```

```python
spearman_corr = df_corr['JobSat_Score'].corr(df_corr['YearsCodePro'], method='spearman')
print("Spearman Correlation:", spearman_corr)
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='YearsCodePro', y='JobSat_Score', data=df_corr, alpha=0.3)
plt.title('Years of Experience vs Job Satisfaction')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Job Satisfaction Score')
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
employment_edu_ct = pd.crosstab(df['EdLevel'], df['Employment'])
```

```python
# Plot as a stacked bar chart
employment_edu_ct.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='Set3')

plt.title('Employment Status by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Employment Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
